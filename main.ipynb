```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_excel("Online_Retail.xlsx")
print(data.head())
print(data.shape)
```

```python
#checked to see if there are any nulls in the dataset
data.isnull().sum()

#drop missing values from columns
data = data.dropna()

#use numerical columns
data_numeric = data[['Quantity', 'UnitPrice']]

#remove any outliers
data_numeric = data_numeric[(data_numeric['Quantity'] > 0) & (data_numeric['UnitPrice'] > 0)]

print(data_numeric.head())
print(data_numeric.shape)
```

## Using Elbow Method
Goal: Find the best number of clusters

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#small sample for speed
data_sample = data_numeric.sample(1000, random_state=42)

#calcualte inertia
inertia = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(data_sample)
    inertia.append(kmeans.inertia_)

#plot
plt.plot(range(1, 11), inertia, marker='o')
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.show()
```

## Applying K-Means Clustering
Goal: Cluster the data

```python
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(data_sample)

#add cluster labels to the sample
data_sample['Cluster'] = clusters

print(data_sample.head())
print("Cluster sizes:", data_sample['Cluster'].value_counts())

```

## Reducing the PCA

```python
from sklearn.decomposition import PCA

#features for PCA
pca_features = data_sample[['Quantity', 'UnitPrice']]

#create PCA object
pca = PCA(n_components=2, random_state=42)
pca_result = pca.fit_transform(pca_features)

#New dataframe
pca_df = pd.DataFrame(pca_result, columns=['PC1', 'PC2'])
pca_df['Cluster'] = data_sample['Cluster']

print(pca_df.head())
print("Explained Variance:", pca.explained_variance_ratio_)

```

### Visualizing and Sharing

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx"
data = pd.read_excel(url)

# Clean data
data = data.dropna()
data_numeric = data[['Quantity', 'UnitPrice']]
data_numeric = data_numeric[(data_numeric['Quantity'] > 0) & (data_numeric['UnitPrice'] > 0)]
data_sample = data_numeric.sample(1000, random_state=42)

# Elbow Method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_sample)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.show()

# K-Means
kmeans = KMeans(n_clusters=3, random_state=42)  # Adjust K
clusters = kmeans.fit_predict(data_sample)
data_sample['Cluster'] = clusters

# PCA
features = data_sample[['Quantity', 'UnitPrice']]
pca = PCA(n_components=2)
pca_result = pca.fit_transform(features)
pca_df = pd.DataFrame(pca_result, columns=['PC1', 'PC2'])
pca_df['Cluster'] = clusters

# Visualize
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=pca_df, palette='deep')
plt.title("PCA of Online Retail Data with K-Means Clusters")
plt.show()

# Info
print("Cluster counts:", data_sample['Cluster'].value_counts())
print("Explained variance ratio:", pca.explained_variance_ratio_)
```
